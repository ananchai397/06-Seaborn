```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 1.สร้าง DataFrame ที่เก็บยอดขายของบริษัท A,B,C,D ของปี 2014,2015,2016,2017 โดยชื่อบริษัทเป็น index และปีเป็น column. 
#สร้างคำสั่ง numpy สุ่มเลขแบบ float ให้เป็นรายได้ของบริษัทที่อยู่ในช่วง 50 – 500 (หน่วยล้านบาท) 

Com  = ["A","B","C","D"]
Year = ["2014","2015","2016","2017"]
Incom = np.random.uniform(50,500,(4,4))
```

```python
df = pd.DataFrame(Incom, Com, Year)
df
```

```python
#2.ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Bar plot ให้แกน x เป็น ชื่อบริษัท ยอดขายเฉลี่ยเป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

M=df.T.mean()
sns.barplot(x=df.index, y=M, data=df)  #แบบแนวตั้ง
```

```python
sns.barplot(x=M, y=df.index, data=df)  #แบบแนวนอน
```

```python
# 3. ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Scatter plot ให้แกน x เป็น ชื่อบริษัท ยอดขายปี 2017 เป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

sns.scatterplot(x=df.index, y=df['2017'], data=df) #แนวตั้ง
```

```python
sns.scatterplot(x=df['2017'], y=df.index, data=df) #แนวนอน
```

```python
#ให้ import ไฟล์ titanic ไฟล์เดิมจากบทเรียน Pandas เข้าสู่ Jupyter Notebook สามารถใช้ Pandas มาร่วมทำด้วย

#4. สร้าง count plot นับข้อมูล Embarked โดยสร้างกราฟแบบแนวตั้งและแนวนอน

df2 = pd.read_csv("train.csv")
df2['Fare'] = pd.to_numeric(df2['Fare'])
df2['Age'] = pd.to_numeric(df2['Age'])

fig = plt.figure(figsize=(10,8))
sns.countplot(x='Embarked', data=df2)  #แนวตั้ง
```

```python
fig = plt.figure(figsize=(10,8))
sns.countplot(y='Embarked', data=df2)  #แนวนอน
```

```python
#5. สร้าง count plot นับข้อมูล Pclass โดยสร้างกราฟแบบแนวตั้งและแนวนอน
fig = plt.figure(figsize=(10,8))
sns.countplot(x='Pclass', data=df2)  #แนวตั้ง
```

```python
fig = plt.figure(figsize=(10,8))
sns.countplot(y='Pclass', data=df2)  #แนวนอน
```

```python
#6. สร้าง box plot ให้แกน X เป็นเพศ และแกน Y เป็นราคา 
fig = plt.figure(figsize=(10,8))
sns.boxplot(x='Sex', y='Fare', data= df2)
```

```python
#7. สร้าง box plot ให้แกน X เป็นPclass และแกน Y เป็นราคา 
fig = plt.figure(figsize=(10,8))
sns.boxplot(x='Pclass', y='Fare', data= df2)
```

```python
#8. สร้าง barplot โดยให้ แกน X เป็นจุดหมายปลายทางและแกน Y เป็นราคา
fig = plt.figure(figsize=(10,8))
sns.barplot(x='Embarked', y='Fare', data= df2)
```

```python
#9. สร้าง barplot โดยให้ แกน X เป็นPclass และแกน Y เป็นอายุ
fig = plt.figure(figsize=(10,8))
sns.barplot(x='Pclass', y='Age', data= df2)
```

```python
#10. สร้าง Stripplot โดยให้แกน X เป็น Pclass และ แกน Y เป็นราคา
fig = plt.figure(figsize=(10,8))
sns.stripplot(x='Pclass', y ="Fare", data=df2)
```

```python
#11. สร้าง Stripplot โดยให้ แกน X เป็น Survived แกน Y เป็นอายุ และ hue เป็นเพศ
fig = plt.figure(figsize=(10,8))
sns.stripplot(x='Survived', y ="Age", data=df2, hue='Sex')
```

```python
#12. สร้าง Swarmplot โดยให้ แกน X เป็นเพศและแกน Y เป็นอายุ
fig = plt.figure(figsize=(10,8))
sns.swarmplot(x='Sex', y='Age',data=df2)
```

```python
#13. สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ 
new_df = pd.crosstab(df2['Pclass'], df2['Sex'])
sns.heatmap(new_df)
```

```python
#14. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา
new_df2 = df2.pivot_table(index='Pclass', columns = 'Sex', values = 'Fare')
sns.heatmap(new_df2, annot = True, cmap='coolwarm')
```

```python
#15. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง
new_df2 = df2.pivot_table(index='Pclass', columns = 'Sex', values = 'Fare')
sns.heatmap(new_df2, annot = True, cmap='coolwarm', linecolor= 'white', linewidth=2)
```

```python
#16. สร้าง Heatmap โดย Correlation ของทั้ง DataFrame

fig = plt.figure(figsize=(10,8))
new_df3 =sns.heatmap(df2.corr(), annot=False, linewidths=2,linecolor='white',cmap='coolwarm')
new_df3
```

```python
#17. สร้าง Clustermap ของทั้ง DataFrame
new_df3= df2.corr()
plt.figure(figsize=(10,8))
sns.clustermap(new_df3)
```

```python
#18. พิจารณาจากข้อมูลที่มีความสัมพันธ์มากที่สุดมาสร้าง Scatterplot
plt.figure(figsize=(10,8))
sns.scatterplot(x=df2['SibSp'], y=df2['Parch'],data = df2)
```

```python
#19. พิจารณาจากข้อมูลที่มีความสัมพันธ์น้อยที่สุดมาสร้าง Scatterplot และให้ hue เป็นเพศ
plt.figure(figsize=(10,8))
sns.scatterplot(x=df2['Pclass'], y=df2['Fare'],data=df2, hue = df2['Sex'])
```
